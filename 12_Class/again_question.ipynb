### 문제 4: **상속의 특성 통합 테스트**

다람쥐는 강사님들의 지도를 받으며 다양한 분야에서 경험을 쌓고 있습니다. 상속과 메서드 오버라이딩, 다중 상속을 종합적으로 테스트하세요.

1. `Mentor` 클래스:
    - 속성 `field`를 생성자로 받아 저장합니다.
    - 메서드 `guide`는 `"저는 [field] 분야의 멘토입니다."`를 반환합니다.
2. `Lecturer` 클래스:
    - 속성 `experience`를 생성자로 받아 저장합니다.
    - 메서드 `teach`는 `"저는 [experience]년의 강의 경력을 가지고 있습니다."`를 반환합니다.
3. `Candidate` 클래스:
    - `Mentor`와 `Lecturer` 클래스를 다중 상속받습니다.
    - `guide` 메서드를 오버라이드하여 `"저는 [field] 분야의 강사이며, 이직 준비를 도와드립니다."`를 반환합니다.

```python
class Mentor:
    def __init__(self, field):
        self.field = field
    def guide(self):
        return f"저는 {self.field} 분야의 멘토입니다."
class Lecturer:
    def __init__(self, experience):
        self.experience = experience
    def teach(self):
        return f"저는 {self.experience} 경력이 있습니다."
    
class Candidate(Mentor, Lecturer):
    def __init__(self, field, experience):
        Mentor.__init__(self, field)
        Lecturer.__init__(self, experience)

    def guide(self):
        return f"{self.field} 이게 호출된다면 정답!"
    

temp12 = Candidate("Ai", 9)
temp13 = Mentor("python")
temp14 = Lecturer(15)
print(temp12.guide())
print(temp13.guide())
print(temp14.teach())
```

문제 2: 기술 강사 후보 평가 시스템

다람쥐는 기술 강사로 지원 중이다. 강사에게는 기술 능력과 강의 능력이 모두 필요하다.
두 부모 클래스를 이용해 다중 상속 기능을 테스트하라.

Developer 클래스

속성 skill 을 생성자로 받는다.

메서드 work() 는 "저는 [skill] 개발을 합니다." 를 반환한다.

Instructor 클래스

속성 topic 을 생성자로 받는다.

메서드 work() 는 "저는 [topic] 강의를 합니다." 를 반환한다.

TechLecturer 클래스

Developer 와 Instructor 를 다중 상속받는다.

생성자는 두 부모의 생성자를 모두 호출해야 한다.

work() 메서드를 오버라이드하여
"저는 [skill] 개발과 [topic] 강의를 모두 할 수 있습니다." 를 반환한다.

```python
class Developer:
    def __init__(self, skill):
        self.skill = skill
    def work(self):
        return f"저는 [{self.skill}] 개발을 합니다."
class Instructor:
    def __init__(self, topic):
        self.topic = topic
    def work(self):
        return f"저는 [{self.topic}] 강의를 합니다."
class TechLecturer(Developer, Instructor):
    def __init__(self, skill, topic):
        Developer.__init__(self, skill)
        Instructor.__init__(self, topic)
    def work(self):
        return f"저는 [{self.skill}] 개발과 [{self.topic}] 강의를 모두 할 수 있습니다."
    
temp17 = TechLecturer("Ai", "Python")
temp18 = Developer("legend")
temp19 = Instructor("of")
print(temp17.work())
print(temp18.work())
print(temp19.work())
```

문제 3: 조직 구성원 관리 시스템

다람쥐 회사는 다양한 직책을 관리한다.
각 직책은 서로 다른 정보를 가지고 있으며,
상속 관계를 가진 클래스를 통해 구성원을 표현하려고 한다.

BaseMember 클래스

속성 name 을 초기화한다.

메서드 info() 는 "이름: [name]" 반환한다.

TeamMember 클래스

BaseMember를 상속받는다.

속성 team 을 초기화한다.

메서드 info() 를 오버라이드하여
"이름: [name], 팀: [team]" 을 반환한다.

Manager 클래스

TeamMember 를 상속받는다.

속성 level 을 초기화한다.

생성자에서는 super()를 꼭 사용해야 한다.

메서드 info() 를 오버라이드하여
"이름: [name], 팀: [team], 직급: [level]" 을 반환한다.

```python
class BaseMember:
    def __init__(self, name):
        self.name = name
    def info(self):
        return f"이름: [{self.name}]"
class TeamMember(BaseMember):
    def __init__(self, name, team):
        self.team = team
        super().__init__(name)
    def info(self):
        return f"이름: [{self.name}], 팀:[{self.team}]"
# Manager 클래스
# TeamMember 를 상속받는다.
# 속성 level 을 초기화한다.
# 생성자에서는 super()를 꼭 사용해야 한다.
# 메서드 info() 를 오버라이드하여
# "이름: [name], 팀: [team], 직급: [level]" 을 반환한다.
class Manager(TeamMember):
    def __init__(self, name, team, level):
        self.level = level
        super().__init__(name, team)
    def info(self):
        return f"이름: [{self.name}], 팀:[{self.team}], 직급: [{self.level}]]"

temp20 = Manager("강승원", "23기", "5레벨")
temp21 = BaseMember("강승원")
temp22 = TeamMember("강승원", "23기")
print(temp20.info())
print(temp21.info())
print(temp22.info())
```
