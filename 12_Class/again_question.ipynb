### **문제 1: 클래스 속성, 인스턴스 속성 테스트**

다람쥐는 이직 준비를 위해 자기소개서를 작성해야 합니다. `Squirrel`이라는 클래스를 만들어 다람쥐의 기본 정보를 설정하세요.

1. 클래스 속성 `species`를 설정하여 "다람쥐"로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `name`과 `current_job`을 설정하세요.
3. 메서드 `introduce`를 만들어 이름과 현재 직업을 소개하는 문자열을 반환하세요.
4. 다른 인스턴스를 생성해도 `species`는 공유되지만, `name`과 `current_job`은 서로 독립적임을 확인하세요.

```python
class Squirrel:
    species = "다람쥐"

    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job
    
    def introduce(self):
        return f"이름은 {self.name}이며, 직업은 {self.current_job}입니다."
    
temp1 = Squirrel("강승원", "Ai 엔지니어")
print(temp1.introduce())
print(temp1.species)
temp2 = Squirrel("홍길동", "률랄")
print(temp2.introduce())
print(temp2.species)
```

### **문제 6: 네임스페이스 테스트 - nonlocal**

다람쥐는 이직 준비 중 추가적인 정보가 필요합니다. 중첩 함수와 `nonlocal` 키워드를 사용해 바깥 함수의 변수를 수정하세요.

1. 함수 `nested_function`을 작성하세요.

1. 바깥 함수의 변수 `status`를 "작성 중"으로 초기화하세요.
2. 중첩 함수에서 `status`를 "완료"로 수정하세요.

```python
def nested_function():
    status = "작성 중"

    def inner_function():
        nonlocal status
        status = "완료"

    print(status)
    inner_function()
    print(status)

nested_function()
```

### 문제 1: **기본 상속을 활용한 클래스 생성**

다람쥐는 이직을 준비하며 새로운 직업군을 탐색하고 있습니다. 다람쥐의 기본적인 정보를 담는 `Person` 클래스를 만들고, 이를 상속받는 `Candidate` 클래스를 생성하세요.

1. `Person` 클래스:
    - 클래스 속성 `national`은 "Korea"로 설정합니다.
    - 메서드 `introduce`는 `"안녕하세요, 저는 한국에서 왔습니다."`를 반환합니다.
2. `Candidate` 클래스:
    - `Person` 클래스를 상속받습니다.
    - 생성자에서 이름(`name`)과 관심 분야(`field`)를 매개변수로 받아 저장합니다.
    - 메서드 `introduce`를 오버라이드하여 `"안녕하세요, 저는 [name]입니다. [field] 분야에 관심이 있습니다."`를 반환합니다.

```python
class Person:
    national = "Korea"
    
    def introduce(self):
        return "안녕하세요, 저는 한국에서 왔습니다."
    
class Candidate(Person):

    def __init__(self, name, field):
        self.name = name
        self.field = field
    def introduce(self):
        return f"안녕하세요, 저는 {self.name}입니다. [{self.field}] 분야에 관심이 있습니다."
    
temp6 = Person()
temp7 = Candidate("강승원", "한국")
print(temp6.introduce())
print(temp7.introduce())
```

### 문제 2: **다중 상속을 활용한 클래스 생성**

다람쥐는 이직을 준비하며 여러 강사님들의 도움을 받고 있습니다. 강사의 역할을 분리한 클래스를 만들어 상속 구조를 테스트하세요.

1. `Mentor` 클래스:
    - 메서드 `guide`는 `"이직 준비를 도와드립니다."`를 반환합니다.
2. `Lecturer` 클래스:
    - 메서드 `teach`는 `"이직에 필요한 기술을 가르쳐드립니다."`를 반환합니다.
3. `Coach` 클래스:
    - `Mentor`와 `Lecturer` 클래스를 다중 상속받습니다.
    - 메서드 `support`는 `"여러 방면으로 지원을 아끼지 않겠습니다."`를 반환합니다.

```python
class Mentor():

    def guide(self):
        return "이직 준비를 도와드립니다."
    
class Lecturer():

    def teach(self):
        return "이직에 필요한 기술을 가르쳐드립니다."
    
class Coach(Mentor, Lecturer):
    
    def support(self):
        return "여러 방면으로 지원을 아끼지 않겠습니다."

temp7 = Mentor()
temp8 = Lecturer()
temp9 = Coach()

print(temp7.guide())
print(temp8.teach())

print(temp9.support())
print(temp9.guide())
print(temp9.teach())
```
