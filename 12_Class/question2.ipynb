다람쥐는 이직 준비 중 클래스를 통해 도구를 관리합니다. `Toolbox` 클래스를 만들어 네임스페이스를 테스트하세요.

1. 클래스 속성 `tools`를 ["노트북", "책상"]으로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `tools`를 설정하고, 객체마다 다른 도구를 가질 수 있도록 설정하세요.
3. `add_tool` 메서드를 만들어 클래스 속성에 새 도구를 추가하세요.

```python
# toolbox.py

class Toolbox:
    # 1) 클래스 속성
    tools = ["노트북", "책상"]

    # 2) 인스턴스 속성(객체마다 다르게)
    def __init__(self, tools):
        self.tools = tools

    # 3) 인스턴스 메서드지만, 클래스 속성에 추가
    def add_tool(self, tool):
        Toolbox.tools.append(tool)   # 또는 self.__class__.tools.append(tool)


# 사용 예시
box1 = Toolbox(["펜"])
box2 = Toolbox(["태블릿"])

box1.add_tool("마우스")   # 클래스 tools에 추가됨

print("클래스 tools:", Toolbox.tools)  # ['노트북', '책상', '마우스']
print("box1 tools:", box1.tools)       # ['펜']
print("box2 tools:", box2.tools)       # ['태블릿']
```

```python
class Mentor:
    def guide(self):
        return "이직 준비 도와드림"
    
class Lecturer:
    def teach(self):
        return "이직 필수 도와드림"

class Coach(Mentor, Lecturer):
    def support(self):
        return "여러 방면으로 지원 아끼지 않음"
    
temp23 = Mentor()
temp24 = Lecturer()
temp25 = Coach()

print(temp23.guide())
print(temp24.teach())
print(temp25.support())
print(temp25.guide())
print(temp25.teach())
```
