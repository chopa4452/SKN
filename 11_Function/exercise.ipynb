1. `work_hours(start, end)`라는 함수를 만들어 매개변수 `start`(출근 시간)와 `end`(퇴근 시간)를 입력받아 근무 시간을 계산하세요.
    - 근무 시간 = `end - start`
2. 반환된 근무 시간을 출력하세요.
    - 출력 예시: **"오늘 근무한 시간은 [근무 시간]시간입니다."**

```python
def work_hours(start, end):

    return end - start

start = int(input(print("몇시에 출근하시나요?")))
end = int(input(print("몇시에 퇴근하시나요?")))

result = work_hours(start, end)

print(f"오늘 근무한 시간은 [{result}시간입니다.]")

```

## **문제 6: 업무 성과 분석하기**

**"다람쥐는 오늘의 업무 성과를 보고해야 한다."**

1. `work_report(tasks_done)`이라는 함수를 만들어 오늘 처리한 업무 수를 매개변수로 받으세요.
2. 함수는 다음 두 값을 반환합니다.
    - `tasks_done`: 처리한 업무 수
    - `status`: 업무 성과 평가 (10개 이상: "우수", 5~9개: "보통", 4개 이하: "노력 필요")
3. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"오늘 처리한 업무는 [업무 수]개입니다. 성과 평가: [평가]"**

```python
def work_report(tasks_done):
    if tasks_done >= 10:
        status = "우수"
    elif 5 <= tasks_done <= 9:
        status = "보통"
    else:
        status = "노력 필요"
    return tasks_done, status

work = int(input("오늘 처리한 업무 수는 몇개인가요?"))
count, grade = work_report(work)
print(f"오늘 처리한 업무는 [{count}]개입니다. 성과 평가: [{grade}]")
```

## **문제 7: 월급 계산 및 세금 공제**

**"다람쥐는 첫 월급을 계산하고 세금을 공제해야 한다."**

1. `calculate_salary(base_salary, bonus)`라는 함수를 만들어 기본 월급과 보너스를 매개변수로 받으세요.
2. 총 월급을 계산하고, 세금은 총 월급의 10%로 공제하세요.
3. 함수는 다음 두 값을 반환합니다.
    - `net_salary`: 실수령액 (총 월급 - 세금)２
    - `tax`: 세금 공제액
4. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"세전 월급: [총 월급]원, 세금: [세금]원, 실수령액: [실수령액]원"**

```python
def calculate_salary(base_salary, bonus):

    incom = base_salary + bonus
    tax = incom * 0.10
    net_salary = incom - tax

    return incom, tax, net_salary

salary = int(input("당신의 기본 월급은 얼마입니까?"))
bonus = int(input("당신의 보너스는 얼마입니까?"))

net_salary, tax, real_salary = calculate_salary(salary, bonus)
print(f"세전 월급: [{net_salary:,}]원, 세금: [{int(tax):,}]원, 실수령액:[{int(real_salary):,}원]")

```
