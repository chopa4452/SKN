```python
# 1.shallow copy
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a)
print(b)
```

```python
a['a'] = 100
print(a)
print(b) # 주소값을 같이 쓰기 때문에 값도 복사가 되는 것이다.
```

```python
# 2.deep copy
c = {'a': 1, 'b': 2, 'c': 3}
d = c.copy()
print(c)
print(d)

```

```python
c['a'] = 300
print(c)
print(d) # 변경되지 않음.
```

```python
c = {'a' : [1,2,3,]}
d = c.copy()

c['a'][0] = 999

print(c)
print(d)
```

```python
a = {'a': [1,2,3], 'b':100}
b = a.copy()

# b = a.copy() => 얕은 복사 - 같은 주소
# deepcopy() => 깊은 복사 - 다른 주소, 완전히 다른 객체를 생성

print(a) #{'a': [1, 2, 3], 'b': 100}
print(b) #{'a': [1, 2, 3], 'b': 100}
```

```python
a['b'] = 200
a['a'][0] = 999
print(a) #{'a': [999, 2, 3], 'b': 200}
print(b) #{'a': [999, 2, 3], 'b': 100}
```

```python
import copy
original = {
    'name' : 'copybara',
    'score' : [100, 90, 80]
}

deep = copy.deepcopy(original)

original['score'][0] = 999

print('original :', original)
print('deep :', deep)
```

### **문제 10**

다음 조건에 맞는 프로그램을 작성하세요.

1. 두 셋 `A = {1, 2, 3, 4}`와 `B = {3, 4, 5, 6}`가 있습니다.
