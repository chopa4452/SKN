```python
import learning_person

class Person():
    national = 'korea'

    def greeting(self):
        return 'Hello. This is python'


class Student(Person, learning_person.Learner):
    pass # 구문을 그냥 종료시키겠다는 명령어

student = Student()
print(student.greeting())
print(student.learn())
```

```python
class Person:
    national = 'korea'

    def greeting(self):
        return 'Hello. This is Python'
    
class Learner(Person):
    def __init__(self, subject):
        self.subject = subject

    def learn(self):
        return 'I am learning ' + self.subject
    
class Student(Learner):
    def __init__(self, name, subject):
        Learner.__init__(self, subject) # 매개변수로 부모클래스로 전달을 하고 Student 클래스에서 전달을 받으려면 부모 클래스의 생성자를 호출해야한다.
        self.name = name

    def greeting(self):
        return "Hello. My name is " + self.name
    
student = Student('카피바라', "Python") 
print(student.greeting()) # greeting()이 부모, 자식클래스에 둘 다 있다면 선언한 클래스가 가장 우선적으로 호출된다.
print(student.learn())    # 그렇기 때문에 Student에 있는 greeting() 이 호출된 것.
```

```python
# 부모클래스 선언
class Person:
    def __init__(self, name, phone=None):
        self.name = name
        self.phone = phone

    # 인스턴스 출력시 실행되는 메서드
    def __str__(self): # 주소값이 안나오고 바로 출력시켜주는 매직 매서드
        return f'{self.name},  {self.phone}'
    
# 자식 클래스 선언
class Employee(Person):
    def __init__(self, name, phone, position, salary):

        # Person.__init__(self, name, phone)
        super().__init__(name, phone)
        self.position = position
        self.salary = salary

p1 = Person('홍길동', '010-1234-5678')
# print(p1.name)
# print(p1)

m1 = Employee('강승원', '010-4782-4452', '직원', '1_000_000_000')
m2 = Employee('나무늘보', '010-1234-3333', '직원', '20_000_000')

# m1의 이름과 직책 출력
print(f"{m1.name}, {m1.position}")
# m1의 인스턴스 호출해서 실행되는 매서드 확인
print(m1)
# m2의 이름과 직책 출력
print(f"{m2.name}, {m2.position}")
# m2의 인스턴스 호출해서 실행되는 매서드 확인
print(m2)
```
