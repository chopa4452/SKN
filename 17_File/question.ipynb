1. `applications.csv` 파일을 생성하고 다음 데이터를 작성하세요.
    
    ```
    회사명,직무,결과

    회사B,소프트웨어 엔지니어,불합격
    ```

2. 파일을 닫고 다시 열어 모든 데이터를 출력하세요.

````python
f = open('aplication1.csv', 'w', encoding='utf-8')
f.write("""회사명,직무,결과
회사A,데이터 분석가,합격
회사B,소프트웨어 엔지니어,불합격
회사C,제품 관리자,진행 중
""")
f.close()

import csv
with open('aplication1.csv', 'r', encoding='utf-8') as csv_file:
    reader = csv.reader(csv_file)
    for row in reader:
        print(", ".join(row)) # join 메서드는 리스트를 문자열로 만들 때 사용
````

````python
# CSV로 형태로 저장하기
import csv

data = [
    ['회사명', "직무", "결과"],
    ['회사A', "데이터 분석가", "합격"],
    ['회사B', "소프트웨어 엔지니어", "불합격"],
    ['회사C', "제품관리자", "진행 중"],
]

with open('application2.csv', 'w', newline="", encoding='utf-8') as f:
    writer = csv.writer(f) # CSV 작성 도구 생성
    writer.writerows(data) # 여러 행을 한꺼번에 작성

# 파일 읽기
with open('aplication1.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f) # csv.reader(객체) : 한 줄씩 리스트 형태로 읽는다.
    for row in reader:
        print(", ".join(row)) # join 메서드는 리스트를 문자열로 만들 때 사용, 리스트를 문자열로 변환하여 출력
````

````python
# dictionary 형태로 저장하기
import csv

data = [
    {'회사명' : "회사A", "직무" : "데이터 분석가", "결과" : "합격"},
    {'회사명' : "회사B", "직무" : "소프트웨어 엔지니어", "결과" : "합격"},
    {'회사명' : "회사C", "직무" : "제품관리자", "결과" : "불합격"}
]

with open('application4.csv', 'w', newline="", encoding='utf-8') as f:
    filename = ["회사명", "직무", "결과"]
    writer = csv.DictWriter(f, fieldnames=filename) # CSV 작성 도구 생성
    writer.writeheader()
    writer.writerows(data) # 여러 행을 한꺼번에 작성

# 파일 읽기
with open('application4.csv', 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f) # csv.reader(객체) : 한 줄씩 리스트 형태로 읽는다.
    for row in reader:
        # print (row)
        print(row) # 딕셔너리 형태로 출력
````

### **문제 3: `.log` 파일 읽기**

다람쥐는 프로그램 실행 로그를 확인하려고 합니다.

1. `system.log` 파일을 다음 내용으로 생성하세요.
    
    ```
    [INFO] 2024-01-01 12:00:00: 시스템 시작
    [WARN] 2024-01-01 12:05:00: 메모리 사용량 증가
    [ERROR] 2024-01-01 12:10:00: 프로그램 종료
    ```
    
2. 파일을 읽어와 `WARN` 또는 `ERROR`가 포함된 줄만 출력하세요.

````python
with open('system1.log', 'w', encoding='utf-8') as f:
    log_content = [
        "[INFO] 2024-01-01 12:00:00: 시스템 시작",
        "[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가",
        "[ERROR] 2024-01-01 12:10:00: 프로그램 종료"
    ]
    f.writelines(log_content)

with open('system.log', 'r', encoding='utf-8') as f:
    for line in f:
        if "WARN" in line or "ERROR" in line:
            print(line)
````
